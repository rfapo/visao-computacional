## ðŸ“ 2.3 RepresentaÃ§Ã£o MatemÃ¡tica de Imagens

### Conceito Fundamental

Uma **imagem digital** Ã© uma representaÃ§Ã£o matemÃ¡tica de uma imagem real, onde cada pixel possui um valor numÃ©rico que representa sua intensidade luminosa.

![RepresentaÃ§Ã£o MatemÃ¡tica de Imagens](https://raw.githubusercontent.com/rfapo/visao-computacional/main/images/modulo2/representacao_matematica_imagens.png)

### CaracterÃ­sticas Fundamentais

#### **1. Matrizes de Pixels**
- Imagens sÃ£o representadas como **matrizes bidimensionais**
- Cada elemento da matriz representa um pixel
- Coordenadas: (linha, coluna) ou (y, x)

#### **2. Profundidade de Bits**
- Determina quantos valores diferentes um pixel pode assumir
- **8-bit**: 256 valores (0-255)
- **16-bit**: 65.536 valores (0-65535)
- **24-bit**: 16.777.216 valores (RGB)

#### **3. ResoluÃ§Ã£o Espacial**
- NÃºmero de pixels por unidade de comprimento
- Medida em DPI (dots per inch) ou PPI (pixels per inch)
- Afeta a qualidade e tamanho da imagem

#### **4. ResoluÃ§Ã£o de Intensidade**
- NÃºmero de nÃ­veis de cinza distintos
- Relacionada Ã  profundidade de bits
- Determina a suavidade das transiÃ§Ãµes

### RepresentaÃ§Ã£o MatemÃ¡tica

```
I(x,y) = f(x,y)
```

Onde:
- **I(x,y)**: Intensidade do pixel na posiÃ§Ã£o (x,y)
- **f(x,y)**: FunÃ§Ã£o que mapeia coordenadas para intensidade

## ðŸŽ¨ 2.4 Tipos de Imagens Digitais

### ClassificaÃ§Ã£o por Canais

![Tipos de Imagens Digitais](https://raw.githubusercontent.com/rfapo/visao-computacional/main/images/modulo2/tipos_imagens_digitais.png)

#### **1. Imagens em Escala de Cinza**

**CaracterÃ­sticas:**
- Cada pixel representa apenas a **intensidade luminosa**
- Valores tÃ­picos: **0 (preto)** a **255 (branco)** para imagens 8-bit
- RepresentaÃ§Ã£o: **Matriz 2D**

**AplicaÃ§Ãµes:**
- AnÃ¡lise mÃ©dica (raios-X)
- Processamento de documentos
- VisÃ£o computacional clÃ¡ssica

#### **2. Imagens Coloridas**

**CaracterÃ­sticas:**
- Cada pixel possui **mÃºltiplos canais**
- **RGB**: Vermelho, Verde, Azul
- **HSV**: Matiz, SaturaÃ§Ã£o, Valor
- RepresentaÃ§Ã£o: **Matriz 3D** (altura Ã— largura Ã— canais)

**Modelos de Cor:**
| Modelo | Componentes | Uso Principal |
|--------|-------------|---------------|
| **RGB** | Red, Green, Blue | Monitores, cÃ¢meras |
| **HSV** | Hue, Saturation, Value | AnÃ¡lise de cor |
| **LAB** | Luminance, A, B | Processamento cientÃ­fico |

#### **3. Imagens BinÃ¡rias**

**CaracterÃ­sticas:**
- Apenas **dois valores**: 0 (preto) e 1 ou 255 (branco)
- Usadas para **segmentaÃ§Ã£o** e anÃ¡lise de formas
- RepresentaÃ§Ã£o: **Matriz 2D** com valores binÃ¡rios

**AplicaÃ§Ãµes:**
- DetecÃ§Ã£o de objetos
- AnÃ¡lise morfolÃ³gica
- Processamento de documentos

## ðŸ”§ 2.5 DemonstraÃ§Ã£o PrÃ¡tica: VisualizaÃ§Ã£o de Tipos de Imagens

Vamos criar e visualizar diferentes tipos de imagens digitais para entender suas caracterÃ­sticas:

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns

def create_sample_images():
    """Cria imagens de exemplo para demonstraÃ§Ã£o"""
    
    # 1. Imagem em escala de cinza
    gray_image = np.zeros((200, 200), dtype=np.uint8)
    
    # Criar gradiente
    for i in range(200):
        gray_image[:, i] = int(255 * i / 199)
    
    # Adicionar formas
    cv2.circle(gray_image, (100, 100), 50, 128, -1)
    cv2.rectangle(gray_image, (50, 50), (150, 150), 200, 2)
    
    # 2. Imagem colorida RGB
    color_image = np.zeros((200, 200, 3), dtype=np.uint8)
    
    # Criar padrÃ£o colorido
    for i in range(200):
        for j in range(200):
            color_image[i, j, 0] = int(255 * i / 199)  # Red
            color_image[i, j, 1] = int(255 * j / 199)  # Green
            color_image[i, j, 2] = int(255 * (i + j) / 398)  # Blue
    
    # Adicionar formas coloridas
    cv2.circle(color_image, (100, 100), 50, (255, 255, 255), -1)
    cv2.rectangle(color_image, (50, 50), (150, 150), (0, 0, 0), 2)
    
    # 3. Imagem binÃ¡ria
    binary_image = np.zeros((200, 200), dtype=np.uint8)
    
    # Criar padrÃ£o binÃ¡rio
    for i in range(0, 200, 20):
        for j in range(0, 200, 20):
            if (i // 20 + j // 20) % 2 == 0:
                binary_image[i:i+20, j:j+20] = 255
    
    # Adicionar formas
    cv2.circle(binary_image, (100, 100), 50, 255, -1)
    cv2.rectangle(binary_image, (50, 50), (150, 150), 0, 2)
    
    return gray_image, color_image, binary_image

def visualize_image_types():
    """Visualiza diferentes tipos de imagens digitais"""
    
    # Criar imagens de exemplo
    gray_img, color_img, binary_img = create_sample_images()
    
    # Configurar visualizaÃ§Ã£o
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    # Imagem em escala de cinza
    axes[0, 0].imshow(gray_img, cmap='gray')
    axes[0, 0].set_title('Imagem em Escala de Cinza\n(200Ã—200, 8-bit)')
    axes[0, 0].axis('off')
    
    # Histograma da imagem em escala de cinza
    axes[1, 0].hist(gray_img.flatten(), bins=50, color='gray', alpha=0.7)
    axes[1, 0].set_title('Histograma de Intensidades')
    axes[1, 0].set_xlabel('Intensidade')
    axes[1, 0].set_ylabel('FrequÃªncia')
    
    # Imagem colorida
    axes[0, 1].imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
    axes[0, 1].set_title('Imagem Colorida RGB\n(200Ã—200Ã—3, 8-bit por canal)')
    axes[0, 1].axis('off')
    
    # Histogramas dos canais RGB
    colors = ['red', 'green', 'blue']
    for i, color in enumerate(colors):
        axes[1, 1].hist(color_img[:, :, i].flatten(), bins=50, 
                        color=color, alpha=0.5, label=f'Canal {color.upper()}')
    axes[1, 1].set_title('Histogramas dos Canais RGB')
    axes[1, 1].set_xlabel('Intensidade')
    axes[1, 1].set_ylabel('FrequÃªncia')
    axes[1, 1].legend()
    
    # Imagem binÃ¡ria
    axes[0, 2].imshow(binary_img, cmap='gray')
    axes[0, 2].set_title('Imagem BinÃ¡ria\n(200Ã—200, 1-bit)')
    axes[0, 2].axis('off')
    
    # Histograma da imagem binÃ¡ria
    axes[1, 2].hist(binary_img.flatten(), bins=2, color='black', alpha=0.7)
    axes[1, 2].set_title('Histograma BinÃ¡rio')
    axes[1, 2].set_xlabel('Valor')
    axes[1, 2].set_ylabel('FrequÃªncia')
    axes[1, 2].set_xticks([0, 255])
    
    plt.tight_layout()
    plt.show()
    
    # AnÃ¡lise estatÃ­stica
    print("=== ANÃLISE ESTATÃSTICA DAS IMAGENS ===")
    print(f"\nImagem em Escala de Cinza:")
    print(f"  - Formato: {gray_img.shape}")
    print(f"  - Tipo: {gray_img.dtype}")
    print(f"  - Valor mÃ­nimo: {gray_img.min()}")
    print(f"  - Valor mÃ¡ximo: {gray_img.max()}")
    print(f"  - Valor mÃ©dio: {gray_img.mean():.2f}")
    
    print(f"\nImagem Colorida:")
    print(f"  - Formato: {color_img.shape}")
    print(f"  - Tipo: {color_img.dtype}")
    print(f"  - Canais: {color_img.shape[2]}")
    
    print(f"\nImagem BinÃ¡ria:")
    print(f"  - Formato: {binary_img.shape}")
    print(f"  - Tipo: {binary_img.dtype}")
    print(f"  - Valores Ãºnicos: {np.unique(binary_img)}")
    
    return gray_img, color_img, binary_img

# Executar demonstraÃ§Ã£o
print("=== DEMONSTRAÃ‡ÃƒO: TIPOS DE IMAGENS DIGITAIS ===")
gray_sample, color_sample, binary_sample = visualize_image_types()
````

## ðŸ” 2.7 DemonstraÃ§Ã£o PrÃ¡tica: OperaÃ§Ãµes BÃ¡sicas

Vamos implementar e visualizar diferentes operaÃ§Ãµes bÃ¡sicas:

````python
def demonstrate_basic_operations():
    """Demonstra operaÃ§Ãµes bÃ¡sicas de processamento de imagem"""
    
    # Criar imagens de exemplo
    img1 = np.zeros((150, 150), dtype=np.uint8)
    img2 = np.zeros((150, 150), dtype=np.uint8)
    
    # Imagem 1: CÃ­rculo
    cv2.circle(img1, (75, 75), 50, 200, -1)
    
    # Imagem 2: RetÃ¢ngulo
    cv2.rectangle(img2, (50, 50), (100, 100), 150, -1)
    
    # OperaÃ§Ãµes aritmÃ©ticas
    add_result = cv2.add(img1, img2)
    sub_result = cv2.subtract(img1, img2)
    
    # OperaÃ§Ãµes lÃ³gicas
    and_result = cv2.bitwise_and(img1, img2)
    or_result = cv2.bitwise_or(img1, img2)
    
    # OperaÃ§Ãµes geomÃ©tricas
    # RotaÃ§Ã£o
    center = (75, 75)
    angle = 45
    scale = 1.0
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(img1, rotation_matrix, (150, 150))
    
    # Escala
    scaled = cv2.resize(img1, (100, 100))
    scaled = cv2.resize(scaled, (150, 150))
    
    # VisualizaÃ§Ã£o
    fig, axes = plt.subplots(3, 3, figsize=(15, 15))
    
    # Imagens originais
    axes[0, 0].imshow(img1, cmap='gray')
    axes[0, 0].set_title('Imagem 1 (CÃ­rculo)')
    axes[0, 0].axis('off')
    
    axes[0, 1].imshow(img2, cmap='gray')
    axes[0, 1].set_title('Imagem 2 (RetÃ¢ngulo)')
    axes[0, 1].axis('off')
    
    axes[0, 2].axis('off')
    
    # OperaÃ§Ãµes aritmÃ©ticas
    axes[1, 0].imshow(add_result, cmap='gray')
    axes[1, 0].set_title('AdiÃ§Ã£o (img1 + img2)')
    axes[1, 0].axis('off')
    
    axes[1, 1].imshow(sub_result, cmap='gray')
    axes[1, 1].set_title('SubtraÃ§Ã£o (img1 - img2)')
    axes[1, 1].axis('off')
    
    axes[1, 2].axis('off')
    
    # OperaÃ§Ãµes lÃ³gicas
    axes[2, 0].imshow(and_result, cmap='gray')
    axes[2, 0].set_title('AND LÃ³gico')
    axes[2, 0].axis('off')
    
    axes[2, 1].imshow(or_result, cmap='gray')
    axes[2, 1].set_title('OR LÃ³gico')
    axes[2, 1].axis('off')
    
    # OperaÃ§Ãµes geomÃ©tricas
    axes[2, 2].imshow(rotated, cmap='gray')
    axes[2, 2].set_title('RotaÃ§Ã£o (45Â°)')
    axes[2, 2].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # AnÃ¡lise quantitativa
    print("=== ANÃLISE QUANTITATIVA DAS OPERAÃ‡Ã•ES ===")
    print(f"\nImagem 1 (CÃ­rculo):")
    print(f"  - Pixels nÃ£o-zero: {np.count_nonzero(img1)}")
    print(f"  - Valor mÃ©dio: {img1.mean():.2f}")
    
    print(f"\nImagem 2 (RetÃ¢ngulo):")
    print(f"  - Pixels nÃ£o-zero: {np.count_nonzero(img2)}")
    print(f"  - Valor mÃ©dio: {img2.mean():.2f}")
    
    print(f"\nAdiÃ§Ã£o:")
    print(f"  - Valor mÃ¡ximo: {add_result.max()}")
    print(f"  - Valor mÃ©dio: {add_result.mean():.2f}")
    
    print(f"\nSubtraÃ§Ã£o:")
    print(f"  - Valor mÃ­nimo: {sub_result.min()}")
    print(f"  - Valor mÃ©dio: {sub_result.mean():.2f}")
    
    return {
        'original1': img1,
        'original2': img2,
        'addition': add_result,
        'subtraction': sub_result,
        'and_op': and_result,
        'or_op': or_result,
        'rotated': rotated
    }

# Executar demonstraÃ§Ã£o
print("=== DEMONSTRAÃ‡ÃƒO: OPERAÃ‡Ã•ES BÃSICAS ===")
operations_results = demonstrate_basic_operations()
````

## ðŸŽ›ï¸ 2.8 TransformaÃ§Ãµes de Intensidade

### Conceito

**TransformaÃ§Ãµes de Intensidade** sÃ£o operaÃ§Ãµes que modificam os valores de pixel de uma imagem para melhorar sua qualidade ou extrair informaÃ§Ãµes especÃ­ficas.

![TransformaÃ§Ãµes de Intensidade](https://raw.githubusercontent.com/rfapo/visao-computacional/main/images/modulo2/transformacoes_intensidade.png)

### Tipos de TransformaÃ§Ãµes

#### **1. TransformaÃ§Ãµes Lineares**

**FÃ³rmula Geral:**
```
g(x,y) = a * f(x,y) + b
```

**ParÃ¢metros:**
- **a**: Controle de contraste
- **b**: Controle de brilho

**Efeitos:**
- **a > 1**: Aumenta contraste
- **a < 1**: Diminui contraste
- **b > 0**: Aumenta brilho
- **b < 0**: Diminui brilho

#### **2. TransformaÃ§Ãµes NÃ£o-Lineares**

**LogarÃ­tmica:**
```
g(x,y) = c * log(1 + f(x,y))
```

**Exponencial:**
```
g(x,y) = c * (exp(f(x,y)) - 1)
```

**PotÃªncia (Gamma):**
```
g(x,y) = c * f(x,y)^Î³
```

#### **3. EqualizaÃ§Ã£o de Histograma**

**Objetivo**: Distribuir uniformemente os nÃ­veis de intensidade

**Processo:**
1. Calcular histograma da imagem
2. Calcular funÃ§Ã£o de distribuiÃ§Ã£o cumulativa
3. Aplicar transformaÃ§Ã£o

**AplicaÃ§Ãµes:**
- Melhoria de contraste
- NormalizaÃ§Ã£o de iluminaÃ§Ã£o
- PrÃ©-processamento para anÃ¡lise

## ðŸ” 2.9 DemonstraÃ§Ã£o PrÃ¡tica: TransformaÃ§Ãµes de Intensidade

Vamos implementar e visualizar diferentes transformaÃ§Ãµes:

````python
def demonstrate_intensity_transformations():
    """Demonstra transformaÃ§Ãµes de intensidade"""
    
    # Criar imagem de exemplo com baixo contraste
    img = np.zeros((200, 200), dtype=np.uint8)
    
    # Criar gradiente com baixo contraste
    for i in range(200):
        for j in range(200):
            img[i, j] = int(50 + 30 * (i + j) / 400)
    
    # Adicionar formas
    cv2.circle(img, (100, 100), 50, 100, -1)
    cv2.rectangle(img, (50, 50), (150, 150), 120, 2)
    
    # TransformaÃ§Ãµes lineares
    # Aumentar contraste
    high_contrast = cv2.convertScaleAbs(img, alpha=2.0, beta=0)
    
    # Diminuir contraste
    low_contrast = cv2.convertScaleAbs(img, alpha=0.5, beta=0)
    
    # Aumentar brilho
    bright = cv2.convertScaleAbs(img, alpha=1.0, beta=50)
    
    # Diminuir brilho
    dark = cv2.convertScaleAbs(img, alpha=1.0, beta=-50)
    
    # TransformaÃ§Ãµes nÃ£o-lineares
    # LogarÃ­tmica
    img_float = img.astype(np.float32)
    log_transformed = np.uint8(255 * np.log(1 + img_float) / np.log(256))
    
    # PotÃªncia (Gamma)
    gamma = 2.0
    power_transformed = np.uint8(255 * np.power(img_float / 255, gamma))
    
    # EqualizaÃ§Ã£o de histograma
    equalized = cv2.equalizeHist(img)
    
    # VisualizaÃ§Ã£o
    fig, axes = plt.subplots(3, 3, figsize=(15, 15))
    
    # Imagem original
    axes[0, 0].imshow(img, cmap='gray')
    axes[0, 0].set_title('Imagem Original\n(Baixo Contraste)')
    axes[0, 0].axis('off')
    
    # Histograma original
    axes[0, 1].hist(img.flatten(), bins=50, color='blue', alpha=0.7)
    axes[0, 1].set_title('Histograma Original')
    axes[0, 1].set_xlabel('Intensidade')
    axes[0, 1].set_ylabel('FrequÃªncia')
    
    axes[0, 2].axis('off')
    
    # TransformaÃ§Ãµes lineares
    axes[1, 0].imshow(high_contrast, cmap='gray')
    axes[1, 0].set_title('Alto Contraste\n(Î±=2.0)')
    axes[1, 0].axis('off')
    
    axes[1, 1].imshow(low_contrast, cmap='gray')
    axes[1, 1].set_title('Baixo Contraste\n(Î±=0.5)')
    axes[1, 1].axis('off')
    
    axes[1, 2].imshow(bright, cmap='gray')
    axes[1, 2].set_title('Brilho Aumentado\n(Î²=50)')
    axes[1, 2].axis('off')
    
    # TransformaÃ§Ãµes nÃ£o-lineares
    axes[2, 0].imshow(log_transformed, cmap='gray')
    axes[2, 0].set_title('TransformaÃ§Ã£o LogarÃ­tmica')
    axes[2, 0].axis('off')
    
    axes[2, 1].imshow(power_transformed, cmap='gray')
    axes[2, 1].set_title('TransformaÃ§Ã£o PotÃªncia\n(Î³=2.0)')
    axes[2, 1].axis('off')
    
    axes[2, 2].imshow(equalized, cmap='gray')
    axes[2, 2].set_title('EqualizaÃ§Ã£o de Histograma')
    axes[2, 2].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # AnÃ¡lise estatÃ­stica
    print("=== ANÃLISE ESTATÃSTICA DAS TRANSFORMAÃ‡Ã•ES ===")
    
    images = {
        'Original': img,
        'Alto Contraste': high_contrast,
        'Baixo Contraste': low_contrast,
        'Brilho Aumentado': bright,
        'LogarÃ­tmica': log_transformed,
        'PotÃªncia': power_transformed,
        'Equalizada': equalized
    }
    
    for name, image in images.items():
        print(f"\n{name}:")
        print(f"  - Valor mÃ­nimo: {image.min()}")
        print(f"  - Valor mÃ¡ximo: {image.max()}")
        print(f"  - Valor mÃ©dio: {image.mean():.2f}")
        print(f"  - Desvio padrÃ£o: {image.std():.2f}")
    
    return images

# Executar demonstraÃ§Ã£o
print("=== DEMONSTRAÃ‡ÃƒO: TRANSFORMAÃ‡Ã•ES DE INTENSIDADE ===")
transformations_results = demonstrate_intensity_transformations()
````
