## 1.2 Evolução Histórica da Visão Computacional### Períodos Históricos**1. Era Pré-Digital (1950s-1970s)**- **Primeiros sistemas**: Análise de padrões simples- **Limitações**: Tecnologia computacional limitada- **Foco**: Reconhecimento de caracteres e formas básicas**2. Era Digital Inicial (1980s-1990s)**- **Digitalização**: Processamento de imagens digitais- **Algoritmos clássicos**: Edge detection, feature extraction- **Aplicações**: OCR, análise médica básica**3. Era Moderna (2000s-2010s)**- **Machine Learning**: Algoritmos de aprendizado- **Features handcrafted**: SIFT, SURF, HOG- **Aplicações**: Reconhecimento facial, análise de vídeo**4. Era Deep Learning (2012-presente)**- **Revolução**: CNNs e deep learning- **ImageNet**: Dataset massivo e competição- **AlexNet**: Primeiro modelo deep learning vencedor![Evolução Histórica](https://raw.githubusercontent.com/rfapo/visao-computacional/main/images/modulo1/evolucao_historica.png)### Marcos Importantes**1957 - Perceptron (Rosenblatt)**- **Primeiro modelo**: De aprendizado neural- **Limitações**: Apenas problemas linearmente separáveis- **Impacto**: Base para redes neurais modernas**1986 - Backpropagation (Rumelhart)**- **Algoritmo**: Para treinar redes neurais- **Revolução**: Possibilitou redes profundas- **Aplicação**: Treinamento eficiente de MLPs**2001 - Viola-Jones**- **Detecção facial**: Algoritmo eficiente- **Features**: Haar-like features- **Impacto**: Primeira detecção facial em tempo real**2004 - SIFT (Lowe)**- **Feature detector**: Scale-Invariant Feature Transform- **Robustez**: Invariante a escala e rotação- **Aplicações**: Matching, panorama, 3D reconstruction**2012 - AlexNet (Krizhevsky et al.)**- **Revolução**: Deep learning para visão computacional- **ImageNet**: Redução de erro de 26% para 15%- **Impacto**: Início da era deep learning**Referências:**- [ImageNet Classification with Deep Convolutional Neural Networks - Krizhevsky et al.](https://papers.nips.cc/paper/2012/hash/c399862d3b9d6b76c8436e924a68c45b-Abstract.html)- [Distinctive Image Features from Scale-Invariant Keypoints - Lowe](https://www.cs.ubc.ca/~lowe/papers/ijcv04.pdf)

## 1.3 O Impacto do ImageNet e AlexNet### ImageNet Dataset**Características do Dataset:**- **14 milhões de imagens**: Dataset massivo- **22.000 categorias**: Organizadas hierarquicamente- **1.000 classes**: Para competição ILSVRC- **1.2 milhões de imagens**: Para treinamento- **50.000 imagens**: Para validação- **100.000 imagens**: Para teste**Estrutura Hierárquica:**- **WordNet**: Organização semântica- **Sinônimos**: Agrupamento de classes similares- **Hierarquia**: Relações entre conceitos![ImageNet Dataset](https://raw.githubusercontent.com/rfapo/visao-computacional/main/images/modulo1/imagenet_dataset.png)### ILSVRC (ImageNet Large Scale Visual Recognition Challenge)**Competição Anual:**- **2010-2017**: Competição principal- **Tarefas**: Classificação, detecção, localização- **Métricas**: Top-1 e Top-5 error rates- **Impacto**: Benchmark para comparação de métodos**Evolução dos Resultados:**- **2010**: 28% error (métodos clássicos)- **2011**: 26% error (melhorias incrementais)- **2012**: 16% error (AlexNet - revolução)- **2013**: 12% error (ZFNet)- **2014**: 7% error (VGG, GoogLeNet)- **2015**: 3.6% error (ResNet)- **2016**: 3% error (melhorias incrementais)- **2017**: 2.3% error (SENet)### AlexNet: A Revolução**Arquitetura Inovadora:**- **8 camadas**: 5 convolucionais + 3 fully connected- **60M parâmetros**: Modelo grande para época- **ReLU activation**: Primeira aplicação em larga escala- **Dropout**: Regularização eficaz- **Data augmentation**: Expansão do dataset- **GPU training**: Paralelização massiva**Inovações Técnicas:**- **ReLU**: Solução para vanishing gradient- **Dropout**: Prevenção de overfitting- **GPU**: Treinamento em paralelo- **Data augmentation**: Rotação, espelhamento, crop![Arquitetura AlexNet](https://raw.githubusercontent.com/rfapo/visao-computacional/main/images/modulo1/arquitetura_alexnet.png)**Impacto Histórico:**- **Redução de erro**: De 26% para 16%- **Revolução**: Início da era deep learning- **Influência**: Base para arquiteturas futuras- **Aplicações**: Transfer learning e fine-tuning**Referências:**- [ImageNet Classification with Deep Convolutional Neural Networks - Krizhevsky et al.](https://papers.nips.cc/paper/2012/hash/c399862d3b9d6b76c8436e924a68c45b-Abstract.html)- [ImageNet: A Large-Scale Hierarchical Image Database - Deng et al.](https://ieeexplore.ieee.org/document/5206848)

## 1.4 Demonstração Prática: Evolução dos Modelos

Vamos visualizar como diferentes técnicas de processamento de imagem evoluíram ao longo do tempo:

### Comparação de Técnicas de Detecção de Bordas

**Objetivo**: Mostrar a evolução das técnicas de detecção de bordas desde métodos clássicos até deep learning.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, feature
import torch
import torchvision.transforms as transforms
from PIL import Image

def load_sample_image():
    """Carrega uma imagem de exemplo para demonstração"""
    # Criar uma imagem sintética com diferentes padrões
    img = np.zeros((200, 200), dtype=np.uint8)
    
    # Adicionar formas geométricas
    cv2.rectangle(img, (50, 50), (100, 100), 255, -1)
    cv2.circle(img, (150, 80), 30, 128, -1)
    cv2.ellipse(img, (100, 150), (40, 20), 45, 0, 360, 200, -1)
    
    # Adicionar ruído
    noise = np.random.normal(0, 10, img.shape).astype(np.uint8)
    img = cv2.add(img, noise)
    
    return img

def demonstrate_edge_detection_evolution():
    """Demonstra a evolução das técnicas de detecção de bordas"""
    
    # Carregar imagem de exemplo
    img = load_sample_image()
    
    # 1. Sobel (1970s) - Primeiro detector de bordas
    sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
    sobel = np.sqrt(sobel_x**2 + sobel_y**2)
    
    # 2. Canny (1986) - Detector mais robusto
    canny = cv2.Canny(img, 50, 150)
    
    # 3. Laplacian of Gaussian (LoG) - Detecção de bordas suavizadas
    log = filters.laplace(img.astype(float))
    
    # 4. Difference of Gaussians (DoG) - Detecção multiescala
    gaussian1 = cv2.GaussianBlur(img, (5, 5), 1)
    gaussian2 = cv2.GaussianBlur(img, (5, 5), 2)
    dog = gaussian1.astype(float) - gaussian2.astype(float)
    
    # Visualização
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    axes[0, 0].imshow(img, cmap='gray')
    axes[0, 0].set_title('Imagem Original')
    axes[0, 0].axis('off')
    
    axes[0, 1].imshow(sobel, cmap='gray')
    axes[0, 1].set_title('Sobel (1970s)')
    axes[0, 1].axis('off')
    
    axes[0, 2].imshow(canny, cmap='gray')
    axes[0, 2].set_title('Canny (1986)')
    axes[0, 2].axis('off')
    
    axes[1, 0].imshow(log, cmap='gray')
    axes[1, 0].set_title('Laplacian of Gaussian')
    axes[1, 0].axis('off')
    
    axes[1, 1].imshow(dog, cmap='gray')
    axes[1, 1].set_title('Difference of Gaussians')
    axes[1, 1].axis('off')
    
    # Espaço para demonstração de deep learning
    axes[1, 2].text(0.5, 0.5, 'Deep Learning\nEdge Detection\n(2010s+)', 
                    ha='center', va='center', fontsize=12, 
                    bbox=dict(boxstyle='round', facecolor='lightblue'))
    axes[1, 2].set_title('Era Moderna')
    axes[1, 2].axis('off')
    
    plt.suptitle('Evolução das Técnicas de Detecção de Bordas', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    return {
        'original': img,
        'sobel': sobel,
        'canny': canny,
        'log': log,
        'dog': dog
    }

# Executar demonstração
edge_results = demonstrate_edge_detection_evolution()
```

### Análise dos Resultados

**Evolução Observada:**

1. **Sobel (1970s)**: Detecção básica de gradientes, mas com ruído
2. **Canny (1986)**: Melhor supressão de ruído e bordas mais limpas
3. **LoG**: Detecção de bordas suavizadas, útil para objetos com bordas suaves
4. **DoG**: Detecção multiescala, captura bordas em diferentes escalas
5. **Deep Learning**: Representação aprendida, adaptativa ao contexto

**Insights Importantes:**
- **Progressão**: De métodos simples para complexos
- **Robustez**: Melhoria na qualidade das bordas detectadas
- **Adaptabilidade**: Deep learning se adapta ao contexto
- **Performance**: Evolução constante da precisão
