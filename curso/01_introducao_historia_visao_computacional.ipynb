## üîç 1.5 Demonstra√ß√£o Pr√°tica: T√©cnicas de Detec√ß√£o de Bordas

Vamos implementar e visualizar diferentes t√©cnicas de detec√ß√£o de bordas usando uma imagem real:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import filters
import os

def load_sample_image():
    """Carrega a imagem de exemplo"""
    
    # Caminho para a imagem do fot√≥grafo
    image_path = "images/modulo1/fotografo.png"
    
    if os.path.exists(image_path):
        # Carregar imagem
        image = cv2.imread(image_path)
        
        # Converter para RGB
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # Converter para escala de cinza
        image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        print(f"Imagem carregada: {image_rgb.shape}")
        return image_rgb, image_gray
    else:
        print(f"Imagem n√£o encontrada em: {image_path}")
        print("Criando imagem sint√©tica para demonstra√ß√£o...")
        
        # Criar imagem sint√©tica
        image_rgb = np.zeros((300, 400, 3), dtype=np.uint8)
        
        # Adicionar formas
        cv2.circle(image_rgb, (150, 100), 50, (255, 0, 0), -1)
        cv2.rectangle(image_rgb, (200, 150), (300, 250), (0, 255, 0), -1)
        cv2.rectangle(image_rgb, (50, 200), (150, 300), (0, 0, 255), -1)
        
        # Adicionar texto
        cv2.putText(image_rgb, 'Sample Image', (100, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        
        # Converter para escala de cinza
        image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
        
        return image_rgb, image_gray

def sobel_edge_detection(image_gray):
    """Detec√ß√£o de bordas usando Sobel"""
    
    # Gradientes Sobel
    sobel_x = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize=3)
    
    # Magnitude do gradiente
    sobel_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
    
    # Dire√ß√£o do gradiente
    sobel_direction = np.arctan2(sobel_y, sobel_x)
    
    # Normalizar para visualiza√ß√£o
    sobel_x_norm = np.uint8(np.absolute(sobel_x))
    sobel_y_norm = np.uint8(np.absolute(sobel_y))
    sobel_magnitude_norm = np.uint8(sobel_magnitude)
    
    return {
        'gradient_x': sobel_x_norm,
        'gradient_y': sobel_y_norm,
        'magnitude': sobel_magnitude_norm,
        'direction': sobel_direction
    }

def canny_edge_detection(image_gray, low_threshold=50, high_threshold=150):
    """Detec√ß√£o de bordas usando Canny"""
    
    # Detec√ß√£o de bordas Canny
    canny_edges = cv2.Canny(image_gray, low_threshold, high_threshold)
    
    return canny_edges

def laplacian_edge_detection(image_gray):
    """Detec√ß√£o de bordas usando Laplaciano"""
    
    # Filtro Laplaciano
    laplacian = cv2.Laplacian(image_gray, cv2.CV_64F)
    
    # Normalizar para visualiza√ß√£o
    laplacian_norm = np.uint8(np.absolute(laplacian))
    
    return laplacian_norm

def log_edge_detection(image_gray, sigma=1.0):
    """Detec√ß√£o de bordas usando LoG (Laplacian of Gaussian)"""
    
    # Aplicar filtro Gaussiano
    gaussian = cv2.GaussianBlur(image_gray, (0, 0), sigma)
    
    # Aplicar Laplaciano
    log = cv2.Laplacian(gaussian, cv2.CV_64F)
    
    # Normalizar para visualiza√ß√£o
    log_norm = np.uint8(np.absolute(log))
    
    return log_norm

def dog_edge_detection(image_gray, sigma1=1.0, sigma2=2.0):
    """Detec√ß√£o de bordas usando DoG (Difference of Gaussians)"""
    
    # Aplicar dois filtros Gaussianos
    gaussian1 = cv2.GaussianBlur(image_gray, (0, 0), sigma1)
    gaussian2 = cv2.GaussianBlur(image_gray, (0, 0), sigma2)
    
    # Diferen√ßa dos Gaussianos
    dog = gaussian1.astype(np.float64) - gaussian2.astype(np.float64)
    
    # Normalizar para visualiza√ß√£o
    dog_norm = np.uint8(np.absolute(dog))
    
    return dog_norm

def visualize_edge_detection_results(image_rgb, image_gray, results):
    """Visualiza resultados das t√©cnicas de detec√ß√£o de bordas"""
    
    fig, axes = plt.subplots(3, 3, figsize=(18, 15))
    
    # Imagem original
    axes[0, 0].imshow(image_rgb)
    axes[0, 0].set_title('Imagem Original')
    axes[0, 0].axis('off')
    
    # Imagem em escala de cinza
    axes[0, 1].imshow(image_gray, cmap='gray')
    axes[0, 1].set_title('Escala de Cinza')
    axes[0, 1].axis('off')
    
    # Sobel - Gradiente X
    axes[0, 2].imshow(results['sobel']['gradient_x'], cmap='gray')
    axes[0, 2].set_title('Sobel - Gradiente X')
    axes[0, 2].axis('off')
    
    # Sobel - Gradiente Y
    axes[1, 0].imshow(results['sobel']['gradient_y'], cmap='gray')
    axes[1, 0].set_title('Sobel - Gradiente Y')
    axes[1, 0].axis('off')
    
    # Sobel - Magnitude
    axes[1, 1].imshow(results['sobel']['magnitude'], cmap='gray')
    axes[1, 1].set_title('Sobel - Magnitude')
    axes[1, 1].axis('off')
    
    # Canny
    axes[1, 2].imshow(results['canny'], cmap='gray')
    axes[1, 2].set_title('Canny')
    axes[1, 2].axis('off')
    
    # Laplaciano
    axes[2, 0].imshow(results['laplacian'], cmap='gray')
    axes[2, 0].set_title('Laplaciano')
    axes[2, 0].axis('off')
    
    # LoG
    axes[2, 1].imshow(results['log'], cmap='gray')
    axes[2, 1].set_title('LoG (Laplacian of Gaussian)')
    axes[2, 1].axis('off')
    
    # DoG
    axes[2, 2].imshow(results['dog'], cmap='gray')
    axes[2, 2].set_title('DoG (Difference of Gaussians)')
    axes[2, 2].axis('off')
    
    plt.tight_layout()
    plt.show()

def analyze_edge_detection_performance(results):
    """Analisa performance das t√©cnicas de detec√ß√£o de bordas"""
    
    print("=== AN√ÅLISE DE PERFORMANCE DAS T√âCNICAS DE DETEC√á√ÉO DE BORDAS ===")
    
    # Calcular estat√≠sticas
    techniques = {
        'Sobel Magnitude': results['sobel']['magnitude'],
        'Canny': results['canny'],
        'Laplaciano': results['laplacian'],
        'LoG': results['log'],
        'DoG': results['dog']
    }
    
    print("\nEstat√≠sticas por T√©cnica:")
    for name, image in techniques.items():
        # Calcular estat√≠sticas
        mean_val = np.mean(image)
        std_val = np.std(image)
        max_val = np.max(image)
        min_val = np.min(image)
        
        # Contar pixels de borda (valores altos)
        edge_pixels = np.sum(image > 128)
        total_pixels = image.size
        edge_percentage = (edge_pixels / total_pixels) * 100
        
        print(f"\n{name}:")
        print(f"  - Valor m√©dio: {mean_val:.2f}")
        print(f"  - Desvio padr√£o: {std_val:.2f}")
        print(f"  - Valor m√°ximo: {max_val}")
        print(f"  - Valor m√≠nimo: {min_val}")
        print(f"  - Pixels de borda (>128): {edge_pixels} ({edge_percentage:.1f}%)")
    
    # Compara√ß√£o qualitativa
    print("\n=== COMPARA√á√ÉO QUALITATIVA ===")
    print("\nCaracter√≠sticas das T√©cnicas:")
    print("\n1. Sobel:")
    print("   - Vantagens: R√°pido, dire√ß√£o do gradiente")
    print("   - Desvantagens: Sens√≠vel ao ru√≠do")
    print("   - Aplica√ß√£o: Detec√ß√£o de bordas em tempo real")
    
    print("\n2. Canny:")
    print("   - Vantagens: Bordas finas, boa precis√£o")
    print("   - Desvantagens: Par√¢metros sens√≠veis")
    print("   - Aplica√ß√£o: Detec√ß√£o precisa de contornos")
    
    print("\n3. Laplaciano:")
    print("   - Vantagens: Simples, detecta mudan√ßas bruscas")
    print("   - Desvantagens: Muito sens√≠vel ao ru√≠do")
    print("   - Aplica√ß√£o: Detec√ß√£o de bordas em imagens suaves")
    
    print("\n4. LoG (Laplacian of Gaussian):")
    print("   - Vantagens: Reduz ru√≠do, bordas suaves")
    print("   - Desvantagens: Computacionalmente caro")
    print("   - Aplica√ß√£o: Detec√ß√£o de bordas em imagens com ru√≠do")
    
    print("\n5. DoG (Difference of Gaussians):")
    print("   - Vantagens: Detecta bordas em m√∫ltiplas escalas")
    print("   - Desvantagens: Par√¢metros complexos")
    print("   - Aplica√ß√£o: Detec√ß√£o de bordas multiescala")

# Executar demonstra√ß√£o
print("=== DEMONSTRA√á√ÉO: T√âCNICAS DE DETEC√á√ÉO DE BORDAS ===")

# Carregar imagem
image_rgb, image_gray = load_sample_image()

# Aplicar t√©cnicas de detec√ß√£o de bordas
print("\nAplicando t√©cnicas de detec√ß√£o de bordas...")

# Sobel
sobel_results = sobel_edge_detection(image_gray)
print("‚úì Sobel aplicado")

# Canny
canny_result = canny_edge_detection(image_gray)
print("‚úì Canny aplicado")

# Laplaciano
laplacian_result = laplacian_edge_detection(image_gray)
print("‚úì Laplaciano aplicado")

# LoG
log_result = log_edge_detection(image_gray)
print("‚úì LoG aplicado")

# DoG
dog_result = dog_edge_detection(image_gray)
print("‚úì DoG aplicado")

# Organizar resultados
results = {
    'sobel': sobel_results,
    'canny': canny_result,
    'laplacian': laplacian_result,
    'log': log_result,
    'dog': dog_result
}

# Visualizar resultados
print("\nVisualizando resultados...")
visualize_edge_detection_results(image_rgb, image_gray, results)

# Analisar performance
analyze_edge_detection_performance(results)

print("\n=== DEMONSTRA√á√ÉO CONCLU√çDA ===")
```

### An√°lise dos Resultados

**Observa√ß√µes Importantes:**

1. **Sobel**:
   - **Gradiente X**: Detecta bordas verticais
   - **Gradiente Y**: Detecta bordas horizontais
   - **Magnitude**: Combina ambos os gradientes
   - **Aplica√ß√£o**: Detec√ß√£o de bordas em tempo real

2. **Canny**:
   - **Precis√£o**: Bordas finas e bem definidas
   - **Robustez**: Boa performance em diferentes condi√ß√µes
   - **Par√¢metros**: Sens√≠vel aos thresholds
   - **Aplica√ß√£o**: Detec√ß√£o precisa de contornos

3. **Laplaciano**:
   - **Simplicidade**: Algoritmo direto
   - **Sensibilidade**: Detecta mudan√ßas bruscas
   - **Ru√≠do**: Muito sens√≠vel ao ru√≠do
   - **Aplica√ß√£o**: Detec√ß√£o de bordas em imagens suaves

4. **LoG (Laplacian of Gaussian)**:
   - **Suaviza√ß√£o**: Reduz ru√≠do antes da detec√ß√£o
   - **Qualidade**: Bordas mais suaves
   - **Custo**: Computacionalmente mais caro
   - **Aplica√ß√£o**: Detec√ß√£o de bordas em imagens com ru√≠do

5. **DoG (Difference of Gaussians)**:
   - **Multiescala**: Detecta bordas em diferentes escalas
   - **Flexibilidade**: Adapt√°vel a diferentes tamanhos
   - **Complexidade**: Par√¢metros mais complexos
   - **Aplica√ß√£o**: Detec√ß√£o de bordas multiescala
